```python
# case text as lowercase, remove punctuation, remove extra whitespace in string and on both sides of string

data_article['remove_lower_punct'] = data_article['Text'].str.lower().str.replace("'", '').str.replace(" \d+", " ").str.replace(' +', ' ').str.strip()
```

```python
import re
```

```python
data_article['remove_lower_punct'] = data_article.apply(lambda row: re.sub(r"http\S+", "", row['remove_lower_punct']), axis=1)
```

```python
data_article['remove_lower_punct'] = data_article.apply(lambda row: re.sub(r"\n", " ", row['remove_lower_punct']), axis=1)
```

```python
data_article['Translated_text']=''
```

```python
data_article
```

```python
import nltk

nltk.download("punkt")
```

```python
for row_index in range(73, 148):
    try:
        clean_text=''

        raw_text=data_article['remove_lower_punct'].iloc[row_index]
        sents = nltk.tokenize.sent_tokenize(raw_text)#we need to break down into sentences long text will not work
        s=1
        for each_sent in sents:
            print('Current row index:',row_index,'current sentnce no',s)
            clean_text+=mt.translate(each_sent, source=dlt.lang.HINDI, target=dlt.lang.ENGLISH)
            s+=1
        data_article['Translated_text'].iloc[row_index]=clean_text
    except:
        print('Row {} not translared...'.format(row_index))
        continue
    data_article.to_csv('/content/drive/MyDrive/Bridgei2i Inter-IIT NLP Competition/Submit_data/evaluation_data_translated.csv')
```
